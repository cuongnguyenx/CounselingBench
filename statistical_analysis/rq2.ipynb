```python
import pandas as pd
import scipy.stats as ss
from collections import Counter
import numpy as np
```

```python
df_gold = pd.read_csv('../data/mct_combined_v3.csv')
categories = ['Overall'] + list(set(df_gold['question_category']))
categories.remove(np.nan)
categories.remove('counseling')
```

```python
df = pd.read_csv('./CounselingQA_metrics_combined.csv')
df.columns
```

```python
def is_medical_model(model_name):
    model_name = str.lower(model_name)
    if 'bio' in model_name or 'med' in model_name or 'asclepius' in model_name or 'clinical' in model_name or 'mentallama' in model_name:
        return True
    return False
```

```python
performance_dict = dict([])
for lbl in categories:
    performance_dict[lbl] = []
for row in df.iterrows():
    if 'zero-shot_None' in row[1]['setting'] and is_medical_model(row[1]['setting']):
        for lbl in categories:
            col = f"Accuracy_{lbl}"
            performance_dict[lbl].append(row[1][col])
```

```python
ss.ttest_rel(performance_dict['counseling skills and interventions'], performance_dict['professional practice and ethics'])
```

```python
pvalues = []
count = 0
for i, lbl1 in enumerate(categories):
    for j, lbl2 in enumerate(categories):
        if i < j:
            print(count)
            count += 1
            print(lbl1)
            print(np.mean(performance_dict[lbl1]))
            print(lbl2)
            print(np.mean(performance_dict[lbl2]))
            print(ss.ttest_rel(performance_dict[lbl1], performance_dict[lbl2]))
            pvalues.append(ss.ttest_rel(performance_dict[lbl1], performance_dict[lbl2]).pvalue)
            print('############################')
```

```python
ss.false_discovery_control(pvalues, method='bh')
```
